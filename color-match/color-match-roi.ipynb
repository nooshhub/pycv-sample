```python
import numpy as np
import cv2 as cv

"""缩小图像,方便看效果
resize会损失像素，造成边缘像素模糊，不要再用于计算的原图上使用
""" 
def resizeImg(src):
    height, width = src.shape[:2]
    size = (800, int(height * 800/width)) 
    img = cv.resize(src, size, interpolation=cv.INTER_AREA)
    return img

"""根据threshold重新计算BGR的值
"""
def bgrWithThreshold(bgr, threshold):     
    
    if type(bgr) == np.ndarray:
        bgr_list = bgr.tolist()
    elif type(bgr) == tuple:
        bgr_list = list(bgr)
    
    newBgr = []
    for x in bgr_list:
        if x + threshold < 0:
            newBgr.append(0)
        elif x + threshold > 255:
            newBgr.append(255)
        else:
            newBgr.append(x + threshold )
    
    return newBgr  

def findPts(mask):
    contours, hierarchy = cv.findContours(mask, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
    print(len(contours))
    listx = []
    listy=[]

    for i in range(0, len(contours)):
        c = contours[i]
        size = cv.contourArea(c)
        print(size)
        if size > 0:
            M = cv.moments(c)
            print(M)
            cX = int(M["m10"] / M["m00"])
            cY = int(M["m01"] / M["m00"])
            listx.append(cX)
            listy.append(cY)

    listxy = list(zip(listx,listy))
    listxy = np.array(listxy)
    return listxy

# 原图
img_white_bg = cv.imread('cmr1.png')
cv.imshow("img_white_bg", resizeImg(img_white_bg))

segment_line_color = (255,0,0)
# 图片里的颜色可能和示例颜色不相等，适当增加点阈值来防色差
threshold = 0
lower = np.array(bgrWithThreshold(segment_line_color, -threshold), dtype="uint8")  
upper = np.array(bgrWithThreshold(segment_line_color, threshold), dtype="uint8")  
# 根据阈值找到对应颜色区域，黑底白块
mask = cv.inRange(img_white_bg, lower, upper) 
cv.imshow("mask", resizeImg(mask))

# 找出所有的轮廓点
idx = cv.findNonZero(mask)
# blackBg = np.full(mask.shape[:2], 0, dtype=np.uint8)
# blackBg = cv.merge((blackBg,blackBg,blackBg))
blackBg = np.zeros(img_white_bg.shape, dtype=np.uint8)
# pts = idx # np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = findPts(mask)
pts = pts.reshape((-1,1,2))
cv.polylines(blackBg,[pts],True,(0,255,255))
cv.imshow("blackBg", resizeImg(blackBg))

# black background
roiMask = np.zeros(mask.shape[:2], np.uint8)
roiMask = cv.fillConvexPoly(roiMask, idx, (255,255,255)) 
cv.imshow("roiMask", resizeImg(roiMask))



cv.waitKey(0)
cv.destroyAllWindows()
```
